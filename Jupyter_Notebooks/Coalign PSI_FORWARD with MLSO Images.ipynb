```python
import pandas as pd
from datetime import datetime, timedelta, timezone
def find_time_of_observation(longitude):
    orbit_data_directory = '/Users/crura/Desktop/Comps/helios_U1DdEhtZg1.lst'
    hi = pd.read_csv(orbit_data_directory,delimiter=r"\s+")
    
    x = np.linspace(228,257,10000000)
    y_knots = hi['HG_LON'][7:-1]
    x_knots = list(range(0,len(y_knots)+1))
    coefs = np.polyfit(hi['DAY'][8:-1], hi['HG_LON'][8:-1], 1)
    y_poly = np.polyval(coefs,x)
#     plt.plot(x,y_poly)
#     plt.scatter(hi['DAY'][8:-1],hi['HG_LON'][8:-1],color='r')
#     plt.xlabel('day')
#     plt.ylabel('Heliographic Longitude')
#     plt.show()

    day_of_observation = x[np.where(y_poly.round(4)==longitude)[0][1]]
    date_time_beginning = datetime(2017, 1, 1, 0, 0, 0, 0,tzinfo=timezone.utc)
    result = date_time_beginning + timedelta(days=day_of_observation-1)
    
    return longitude,result.isoformat()+'Z'
    

```
