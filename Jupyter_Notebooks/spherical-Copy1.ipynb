```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sci
from tqdm import tqdm_notebook
df = pd.read_csv('/Volumes/Seagate Backup Plus Drive 1/Chris/Magnetic_Field/eclipse2017_mhd_final/density.csv',header=None)

x = df.to_numpy()

density_reshape = x.reshape(294,314,698)
rho = density_reshape
r = pd.read_csv('/Volumes/Seagate Backup Plus Drive 1/Chris/Magnetic_Field/eclipse2017_mhd_final/r.csv',header=None)
t = pd.read_csv('/Volumes/Seagate Backup Plus Drive 1/Chris/Magnetic_Field/eclipse2017_mhd_final/t.csv',header=None)
p = pd.read_csv('/Volumes/Seagate Backup Plus Drive 1/Chris/Magnetic_Field/eclipse2017_mhd_final/p.csv',header=None)

rnew = r.to_numpy()
tnew = t.to_numpy()
pnew = p.to_numpy()

xyz_step = 0.05 # Rs
#xyz_min = -15.0 # Rs
#xyz_max = +15.0 # Rs
xyz_min = -15.0 # Rs
xyz_max = 15.0 # Rs

N = (xyz_max - xyz_min)/xyz_step + 1

```

```python
rho_xyz = np.zeros((int(N),int(N),int(N))) # create empty N x N x N array
n_xyz = np.zeros((int(N),int(N),int(N))) # create empty N x N x N array

r_min = 1.1
r_max = xyz_max





wr = np.where((rnew > r_min) & (rnew > r_min))#rnew[(rnew > r_min) & (rnew < r_min)]#np.where((r > r_min) & (r > r_min))# and np.where((r > r_min)) 
Nr = len(rnew[wr])
Nt = len(tnew)
Np = len(pnew)


for i in tqdm_notebook(range(15,Nr-1)):
    if i % 10 == 0:
        print((100*i)/Nr)
    for j in range(0,Nt-1):
        for k in range(0,Np-1):
            
            x = rnew[[i]]*np.sin(tnew[j])*np.cos(pnew[k])
            y = rnew[[i]]*np.sin(tnew[j])*np.sin(pnew[k])
            z = rnew[[i]]*np.cos(tnew[j])
            
            
            x_ = int((x - xyz_min)/xyz_step)
            y_ = int((y - xyz_min)/xyz_step)
            z_ = int((z - xyz_min)/xyz_step)
            
            
            rho_xyz[x_,y_,z_] = rho_xyz[x_,y_,z_] + rho[i,j,k]
            n_xyz[x_,y_,z_] = n_xyz[x_,y_,z_] + 1
```
